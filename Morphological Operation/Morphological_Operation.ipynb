## 2) Morphological Operation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Task_3.jpg')
```

#### Erosion function:

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#cv2.imwrite("Task_3.jpg",img)
```

#### Erosion Operation:

```python
erosion = cv2.erode(img,kernel,iterations = 2)
plt.imshow(cv2.cvtColor(erosion, cv2.COLOR_BGR2RGB))
cv2.imwrite("Task_3_erosion.jpg",erosion)
```

####  Dilation Operation:

```python
dilation = cv2.dilate(img,kernel,iterations = 2)
plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_BGR2RGB))
cv2.imwrite("Task_3_dilation.jpg",dilation)
```

#### Opening Operation:

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel,iterations = 2)
plt.imshow(cv2.cvtColor(opening, cv2.COLOR_BGR2RGB))
cv2.imwrite("Task_3_opening.jpg",opening)
```

#### Closing Operation:

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel,iterations = 2)
plt.imshow(cv2.cvtColor(closing, cv2.COLOR_BGR2RGB))
cv2.imwrite("Task_3_closing.jpg",closing)
```

#### Morphological Gradient Operation:

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel,iterations = 2)
plt.imshow(cv2.cvtColor(gradient, cv2.COLOR_BGR2RGB))
cv2.imwrite("Task_3_gradient.jpg",gradient)
```
